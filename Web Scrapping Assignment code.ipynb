```python
import pandas as pd
import requests,webbrowser 
from bs4 import BeautifulSoup
import numpy as np
```

```python
data = pd.read_excel("Web Scraping Assignment.xlsx")
```

```python
data.head()
```

```python
df = data[72:]
```

```python

responseDict = {}
df.reset_index(drop=True, inplace=True)
for idx, i in enumerate(df["COLLEGE NAME"]):
    try:
        
        url = str("https://www.google.com/search?q= "+i+"google Address")
        webpage=requests.get(url).text
        soup = BeautifulSoup(webpage, "html.parser")
        address = soup.find_all('div','BNeawe iBp4i AP7Wnd')[0].get_text()
        print(address)
        responseDict[i] = address
#         df["ADRESS"].iloc[idx] = address
        
#         blank_df.append({"COLLEGE NAME":str(i),"ADRESS":str(address)},ignore_index=True)
        
    except:        
        print("not found")
#         df["ADRESS"].iloc[idx] = np.nan
#     break
        
```

```python
data.reset_index(drop=True, inplace=True)
for idx, d in enumerate(data["COLLEGE NAME"]):
#     print(d)
    if d in responseDict:
        print(d)
        data["ADRESS"].iloc[idx] = responseDict[d]
```
