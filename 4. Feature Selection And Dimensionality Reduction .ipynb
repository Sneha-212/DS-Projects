### Importing Libraries

```python
#Importing importatnt libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings("ignore")
```

```python
#Loading the dataset
data=pd.read_csv("DS C4-Project/Problem Statement - Dimentionality Reduction-Dataset.csv")
```

```python
data.head()
```

### Data Preprocessing

```python
#Dropping the product id variable
data=data.drop("Product_ID",axis=1)
```

```python
#shape of data
data.shape
```

```python
#checking the type of each variable
data.dtypes
```

```python
#percentage of missing values in each variable of data
a=data.isna().sum()/len(data)*100
a
```

```python
#saving columns in a variable
variables=data.columns
```

```python
#new variable to store varibles having missing values less than a threshold
variable=[]
for i in range(data.columns.shape[0]):
    #setting the threshold at 60%
    if a[i]<=60:
        variable.append(variables[i])
```

```python
variable
```

```python
#creating the new dataframe using these variable
new_data=data[variable]
new_data.head()
```

```python
#percentage of missing values of each variable in new data
new_data.isna().sum()/len(new_data)*100
```

```python
#imputing the missing values with with the median of product category 2
new_data=new_data.fillna(new_data["Product_Category_2"].median())
```

```python
#checking the missing values
new_data.isna().sum()/len(new_data)*100
```

```python
#Importing label encoder for converting the categorical values to numeric value
from sklearn.preprocessing import LabelEncoder
```

```python
#Transforming each feature
le=LabelEncoder()
label_gender=le.fit_transform(new_data["Gender"])
label_age=le.fit_transform(new_data["Age"])
label_city_category=le.fit_transform(new_data["City_Category"])
label_Stay_In_Current_City_Years=le.fit_transform(new_data["Stay_In_Current_City_Years"])
```

```python
#storing the generated label in their respective column
new_data["Gender"]=label_gender
new_data["Age"]=label_age
new_data["City_Category"]=label_city_category
new_data["Stay_In_Current_City_Years"]=label_Stay_In_Current_City_Years
```

```python
new_data.head()
```

### Q1.Check the correlation matrix and identify most correlated features.

```python
#Dropping the target variable
df=new_data.drop("Purchase",axis=1)
```

```python
#calculating the correlation between indepedent variable
df.corr()
```

```python
#creating correlation matrix
corr_matrix=df.corr().abs()
```

```python
#selecting upper triangle of correlation matrix
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
```

```python
upper
```

```python
#finding index of variables with correlation greater than 0.3
high_corr=[column for column in upper.columns if any(upper[column]>0.3)]
high_corr
```

**After checking the correlation matrix, the most correlated features are Marital_Status', 'Product_Category_2'.**

### Q2.Find out the covariance matrix ,Also identify eigen values and eigen vectors of covariance matrix.

```python
#checking the preprocessed data
new_data.head()
```

```python
#standardise the data
scaler=StandardScaler()
new_data_scale=scaler.fit_transform(new_data)
```

```python
#creating the dataframe
new_data_scale_df=pd.DataFrame(new_data_scale,columns=new_data.columns)
```

```python
#finding the covariance matrix
cov_matrix=new_data_scale_df.cov()
```

**Covariance Matrix**

```python
cov_matrix
```

**Finding Eigen value and Eigen vectors of covariance matrix**

```python
#storing the eigen value and eigen vectors using np.linalg.eig() of the above covariance matrix
eig_val,eig_vect=np.linalg.eig(cov_matrix)
```

**EigenValues**

```python
eig_val=pd.DataFrame(eig_val,index=cov_matrix.columns,columns=["eigen_value"])
eig_val
```

```python
eig_vect=pd.DataFrame(eig_vect,index=cov_matrix.columns)
eig_vect
```

### Q3.Find out the two most important Principal Components.

```python
#Taking the absolute values of eigenvalues
eig_val_magnitude=eig_val.abs()
```

```python
#sorting the values in it's increasing order
eig_val_magnitude=eig_val_magnitude.sort_values("eigen_value",ascending=False)
```

```python
eig_val_magnitude
```

```python
#total sumof eigen values
addition=sum(eig_val_magnitude["eigen_value"])
```

```python
#Finding the contributio of each eigen value by iterating and storing them in a list var_percent
var_percent=[]
for i in range(len(eig_val_magnitude)):
    a=eig_val_magnitude["eigen_value"][i]/addition
    var_percent.append(a)
```

```python
#creating a dataframe
var_percent=pd.DataFrame(var_percent,index=cov_matrix.columns,columns=["%contribution"])
```

```python
var_percent
```

**From the above values of variance obtained,we can observe that two of the most Principal Components are User_ID and "Gender".**

### Q4.Select only 8 features using PCA.

```python
#checking the scaled data
new_data_scale_df.head()
```

```python
#segregating the feature and target variables
X=new_data_scale_df.drop("Purchase",axis=1)
y=new_data_scale_df["Purchase"]
```

```python
#creating the training and validation set

x_train,x_valid,y_train,y_valid=train_test_split(X,y,random_state=10,test_size=0.25)
```

```python
#Fitting the PCA algorithm on our data to find appropriate number of principal components to keep
pca=PCA().fit(x_train)
```

```python
#Plotting the cumulative summation of the Explained Variance
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
plt.xlabel("Number of Components")
plt.ylabel("Variance(%)")
plt.title("Explained Variance")

#We want to preserve 95% of the variance ,so plot a horizontal line at 95% variance
plt.axhline(y=95,color="r",linestyle="--")
plt.show()
```

```python
#Transforming the original train and validation set to 8 components
#Fit on the train data
pca=PCA(8).fit(x_train)

#Transform both training and validation data
x_train_transformed=pca.transform(x_train)
x_valid_transformed=pca.transform(x_valid)
```

```python
x_train_transformed.shape
```

```python
#creating random forest regressor model
model=RandomForestRegressor(random_state=1,max_depth=3,n_estimators=100)
```

```python
#Fitting the same random forest model on same data
model.fit(x_train_transformed,y_train)
```

```python
#Predicting for the transformed validation data
pred2=model.predict(x_valid_transformed)
```

```python
#checking the training performance after transformation
mse(model.predict(x_train_transformed),y_train)
```

```python
#checking the validation performance after transformation
mse(pred2,y_valid)
```
